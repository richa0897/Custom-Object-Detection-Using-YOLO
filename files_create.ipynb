### Importing Libraries

```python
import shutil
import os
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

### Splitting the data into train and test sets

```python
path=os.getcwd()
print(path)
src=os.path.join(path,"Auto")
print(src)
allFileNames=os.listdir(src)
#print(allFileNames)
```

```python
test_ratio=0.3 #split ratio
np.random.shuffle(allFileNames)
train_FileNames,test_FileNames = np.split(np.array(allFileNames),[int(len(allFileNames)* (1 - test_ratio))]) #randomly selecting files for training and testing

train_list=[]
test_list=[]
for name in train_FileNames.tolist():
    path=os.path.join(src,name)
    #print(path)
    train_list.append(path)
for name in test_FileNames.tolist():
    path=os.path.join(src,name)
    #print(path)
    test_list.append(path)
print(len(train_FileNames))
print(len(test_FileNames))
print(train_list[0])
```

```python
src=os.getcwd()
src=os.path.join(src,'Auto')
print(src)
os.makedirs(src+'/images/')
path=os.path.join(src,'images')
print(path)
os.makedirs(path +'/train/')
os.makedirs(path +'/test/')
trainsrc=os.path.join(path,"train")
print(trainsrc)
testsrc=os.path.join(path,"test")
print(testsrc)

#copying train and test files to /images/train and /images/test directory
for name in train_list:
    print(name)
    shutil.copy(name, trainsrc)

for name in test_list:
    print(name)
    shutil.copy(name, testsrc)
```
