### Creating the custom Yaml file 

```python
#directory structure
#Input
#   --datasetv1
#       --images
#            --train
#            --test
#       --labels
#            --train
#            --test

#Output
#   --kaggle/working
#       --yolov5
#       --auto.yaml



cwd = '/kaggle/working'
classes=['Auto']

data = dict(
    train = '/kaggle/input/datasetv1/images/train' ,
    val  = '/kaggle/input/datasetv1/images/test',
    nc    = 1,
    names = classes
    )

with open(os.path.join( cwd , 'auto.yaml'), 'w') as outfile:
    yaml.dump(data, outfile, default_flow_style=False)
```
